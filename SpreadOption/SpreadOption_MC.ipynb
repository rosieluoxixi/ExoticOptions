```python
import pandas as pd
import numpy as np
from numpy import random
```

```python
def SpreadOption_MC(s1, s2, vol1, vol2, rho, strike, nsim=10000, T=1, r=0.03):
    drift1, drift2 = r-0.5*vol1**2*T, r-0.5*vol2**2*T
    diff1, diff2 = vol1*np.sqrt(T), vol2*np.sqrt(T)
    df = np.exp(-r*T)
    mean = (0, 0)
    cov = [[1, rho], [rho, 1]]
    z1, z2 = np.random.multivariate_normal(mean, cov, nsim).T
    s1_T, s2_T = s1*np.exp(drift1+diff1*z1), s2*np.exp(drift2+diff2*z2)
    payoff = np.maximum(s1_T-s2_T-strike*s1, 0)
    price = df*np.mean(payoff)
    return price
```

```python
def TestSpreadOption_MC(testdf):
    OptionPrices = []
    for i in range(0, testdf.shape[0]):
        s1, s2, vol1, vol2, rho, strike, OptionPrice = testdf.iloc[i]
        vol1, vol2, strike = float(vol1.strip('%'))/100, float(vol2.strip('%'))/100, float(strike.strip('%'))/100
        OptionPrice = SpreadOption_MC(float(s1), float(s2), vol1, vol2, float(rho), strike, nsim=10000, T=1, r=0.03)
        OptionPrices.append(OptionPrice)
    return OptionPrices
```
