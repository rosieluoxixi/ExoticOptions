```python
import pandas as pd
import numpy as np
from scipy.stats import norm
```

```python
def SpreadOption_Kirk(s1, s2, vol1, vol2, rho, strike, T=1, r=0.03):
    df = np.exp(-r*T)
    factor = s2/(s2+strike*s1*df)
    local_vol = np.sqrt(vol1**2 + vol2**2*factor**2-2*rho*vol1*vol2*factor)*np.sqrt(T)
    d1 = np.log(s1/(s2+strike*s1*df))/local_vol + 0.5*local_vol
    d2 = d1 - local_vol
    price = s1*norm.cdf(d1) - (s2+strike*s1*df)*norm.cdf(d2)
    return price
```

```python
def TestSpreadOption_MC(testdf):
    OptionPrices = []
    for i in range(0, testdf.shape[0]):
        s1, s2, vol1, vol2, rho, strike, OptionPrice = testdf.iloc[i]
        vol1, vol2, strike = float(vol1.strip('%'))/100, float(vol2.strip('%'))/100, float(strike.strip('%'))/100
        OptionPrice = SpreadOption_Kirk(float(s1), float(s2), vol1, vol2, float(rho), strike, T=1, r=0.03)
        OptionPrices.append(OptionPrice)
    return OptionPrices
```
